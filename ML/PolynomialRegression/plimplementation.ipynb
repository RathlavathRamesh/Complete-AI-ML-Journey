```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
x=6*np.random.rand(100,1) -3
y=0.5*x**2+1.5*x+2+np.random.randn(100,1)
# this is my quadratic equation where a=0.5, b=1.5, c=2
# 0.5*X^2+1.5*X+2
```

```python
plt.scatter(x,y,color="green")
plt.xlabel("")
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
regression1=LinearRegression()
```

```python
regression1.fit(x_train,y_train)
```

```python
print(regression1.coef_)
print(regression1.intercept_)
```

```python
from sklearn.metrics import r2_score
```

```python
score=r2_score(y_test,regression1.predict(x_test))
print(score)
```

```python
plt.plot(x_train,regression1.predict(x_train),color="red")
plt.scatter(x_train,y_train)
plt.xlabel("X Dataset")
plt.ylabel("Y values")
```

```python
# we can clearly see that the single line plot is not fitting the data properly 
# so here we have to use the multiple degree equation to solve the problem 
```

```python
#LETS APPLY POLYNOMIAL TRANSFORMATION 
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly=PolynomialFeatures(degree=2,include_bias=True)
x_train_poly=poly.fit_transform(x_train)
x_test_poly=poly.transform(x_test)
```

```python
x_train_poly
```

```python
x_test_poly
```

```python
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import r2_score
```

```python
regression2=LinearRegression()
regression2.fit(x_train_poly,y_train)
y_predict=regression2.predict(x_test_poly)
score=r2_score(y_test,y_predict)
print(score)
```

```python
print(regression2.coef_)
print(regression2.intercept_)
```

```python
plt.scatter(x_train,regression2.predict(x_train_poly))
plt.scatter(x_train,y_train)
```

```python
poly=PolynomialFeatures(degree=3,include_bias=True)
x_train_poly=poly.fit_transform(x_train)
x_test_poly=poly.transform(x_test)
```

```python
x_train_poly
```

```python
regression3=LinearRegression()
regression3.fit(x_train_poly,y_train)
y_predict=regression3.predict(x_test_poly)
score=r2_score(y_test,y_predict)
print(score)
```
