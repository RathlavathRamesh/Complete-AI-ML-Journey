# flight price dataset includes of various features like airline, source, destination, duration, total stops, date of journey, month of journey, day of journey, arrival time, departure time and price.

```python
import pandas as pd 
```

```python
df=pd.read_excel('flight_price.xlsx')
df 
```

```python
# get the basin info about the data 
df.info()
```

# Feature Engineering Process


```python
df['date']=df['Date_of_Journey'].str.split('/').str[0]
df['month']=df['Date_of_Journey'].str.split('/').str[1]
df['year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
# the problem with this is that if we see the df.info() the date, month, year are object data types we need to convert them into integer data type
```

```python
df['date']=df['date'].astype(int)
df['month']=df['month'].astype(int)
df['year']=df['year'].astype(int)

df.info()
# now we can see that the date, month, year are converted into integer data type
```

```python
# Now we can drop the Date_of_Journey column as we have extracted the date, month, year from it
df.drop('Date_of_Journey',axis=1,inplace=True)
df.head()
```

```python
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df['Arrival_hour']=df['Arrival_Time'].str.split(':').str[0]
df['Arrival_minute']=df['Arrival_Time'].str.split(':').str[1]
df.head()
```

```python
df['Arrival_hour']=df['Arrival_hour'].astype(int)
df['Arrival_minute']=df['Arrival_minute'].astype(int)
# we can drop the Arrival_Time column as we have extracted the hour and minute from it
df.drop('Arrival_Time',axis=1,inplace=True)
df.info()
```

```python
# Now we have to do the same process for the Dep_Time column
df['Dep_hour']=df['Dep_Time'].str.split(':').str[0]
df['Dep_minute']=df['Dep_Time'].str.split(':').str[1]
df.head()
```

```python
df.info()
# now we can remove the Dep_Time column as we have extracted the hour and minute from it
df.drop('Dep_Time',axis=1,inplace=True)

df['Dep_hour']=df['Dep_hour'].astype(int)
df['Dep_minute']=df['Dep_minute'].astype(int)

df.info()
```

```python
df.head()
```

# Now Lets Do the Categorical Features 
# Total_Stops, Airline, Source, Destination, Route, Additional_Info

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':1})
```

```python
df.isna()
```

```python
# Now we have to handle the Duration column
df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['hour']=df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['minute']=df['Duration'].str.split(' ').str[1].str.split('m').str[0]

df.head(2)

df['hour'].isna()
df['minute'].isna()
```

# Now we have to Convert all the categorical columns into numerical columns
# we can use the OneHotEncoder for this


```python
from sklearn.preprocessing import OneHotEncoder

encoder=OneHotEncoder()
```

```python
pd.DataFrame(encoder.fit_transform(df[['Airline','Source','Destination']]).toarray(),columns=encoder.get_feature_names_out())
```

```python
# now we have to merge this dataframe with the original dataframe
ohe_df=pd.DataFrame(encoder.fit_transform(df[['Airline','Source','Destination']]).toarray(),columns=encoder.get_feature_names_out())

df=pd.concat([df,ohe_df],axis=1)

df.head()
```
